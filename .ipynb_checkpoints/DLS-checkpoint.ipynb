```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F','G'],
    'D':['H','I'],
    'E':['J','K'],
    'F':['L','M'],
    'G':['N','O'],
    'H':[],
    'I':[],
    'J':[],
    'K':[],
    'L':[],
    'M':[],
    'N':[],
    'O':[]
}

def DLS(start,goal,path,level,maxD):
  print('\nCurrent level--> ',level)
  print('Goal node testing for',start)
  path.append(start)
  if start == goal:
    print("Goal test successful")
    return path
  print('Goal node testing failed')
  if level==maxD:
    return False
  print('\nExpanding the current node',start)
  for child in graph[start]:
    if DLS(child,goal,path,level+1,maxD):
      return path
    path.pop()
#     print(path)
  return False
  
  
  
start = 'A'
goal = input('Enter the goal node:-')
maxD = int(input("Enter the maximum depth limit:-"))
print()
path = list()
res = DLS(start,goal,path,0,maxD)
if(res):
    print("\nPath to goal node available")
    print("Path : ",path)
else:
    print("No path available for the goal node in given depth limit")
            
```

```python
road = []
visited = set()
def DLS2(start, goal, level, Mdepth):
    if level <= Mdepth:
        print(f"current level : {level}")
        print(f"testing for {start}")
        road.append(start)
        visited.add(start)
        if start == goal :
            print("Destination found ")
            return road
        else:
            print("Still searching.....")
            
        for child in graph[start]:
            if child not in visited:
                if DLS2(child, goal, level+1, Mdepth):
                    return road
        return False
    else:
        return False

n = int(input("Enter max depth : "))
go = input("\nEnter goal node : ")

if DLS2('A', go, 0, n):
    print(f"\nPath found : {road}")
else:
    print(f"\nPath not found for given depth limit")
        
        
        
```
